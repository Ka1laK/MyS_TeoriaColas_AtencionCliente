## TIEMPO DE ATENCION

```python
import numpy as np
import pandas as pd

min_val = 3
mode_val = 7.05
max_val = 15

data = np.random.triangular(min_val, mode_val, max_val, 2000)
df = pd.DataFrame(data, columns=['GeneratedData'])
df.to_csv('tiempo de atencion.csv', index=False)
df.to_csv('tiempo_atencion.txt', index=False, header=False, sep='\t', float_format='%.5f')
df
```

## TIEMPO DE ESPERA DE RESPUESTA

```python
min_val = 0
max_val = 752

data = np.random.uniform(min_val, max_val, 2000)
df2 = pd.DataFrame(data, columns=['GeneratedData'])
df2.to_csv('tiempo de espera de respuesta.csv', index=False)
df2
```

## TIEMPO ENTRE LLEGADAS DE SOLICITUDES

```python
scale_param = 113
data = np.random.exponential(scale_param, 2000)
df3= pd.DataFrame(data, columns=['GeneratedData'])
df3.to_csv('tiempo entre llegadas de solicitudes.csv', index=False)
df3
```
